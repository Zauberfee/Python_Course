#### 1. Count words in a text

a. Output a list of words in the file along with their frequency counts (ignoring case).   
a. Count how many unique words there are (ignoring case).    
c. Check how common are all different sequences of vowels (e.g. the sequences "ieu" or just "e" in "lieutenant")?

```python
# a)
import re


def prep_text(p_text):
    p_text = p_text.lower()

    p_text = p_text.replace("\n", " ")
    

    p_text = re.sub("[\d,\"\.$;]", "", p_text)

    p_text = p_text.replace("&ampamp", "")
    
    p_text = re.sub("\s+", " ", p_text)
    
    l_words = p_text.split(" ")
    
    return l_words


def create_unique_list(p_list):
    l_finished = False
    l_loop_count = 0
    l_unique_words = []

    while not l_finished:
        l_unique_words.append((p_list.count(p_list[0]), p_list[0]))
    
        l_remove_word = p_list[0]
    
        try:
            while True and len(p_list) > 0:
                p_list.remove(l_remove_word)
        except ValueError:
            pass
    
        if(len(p_list) == 0):
            l_finished = True
      
    return l_unique_words

def print_list(p_list):
    l_loop_count_1 = 0
        
    for result in p_list:
        print(p_list[l_loop_count_1])
        l_loop_count_1 = l_loop_count_1 + 1
        
        
v_word_list = prep_text(v_short_text)
v_unique_list = create_unique_list(v_word_list)
print_list(v_unique_list)

          
```

```python
# c)

def select_vowels(p_text):
    l_vowels = re.findall("[aeiou]+", p_text)
    l_vowelcombos = []
    
    l_loop_count = 0
    for result in l_vowels:
        l_vowelcombos.append(l_vowels[l_loop_count])
        
        l_loop_count = l_loop_count + 1
        
    return l_vowelcombos     


v_vowels = select_vowels(v_short_text)
v_unique_vowels = create_unique_list(v_vowels)
print_list(v_unique_vowels)
```

#### 2. Sorting and reversing lines of text

a. Sort each line alphabetically (ignoring case).  
b. Sort in numeric ([ascii](https://python-reference.readthedocs.io/en/latest/docs/str/ASCII.html)) order.  
c. Alphabetically reverse sort (ignoring case).  
d. Sort in reverse numeric ([ascii](https://python-reference.readthedocs.io/en/latest/docs/str/ASCII.html)) order.  

```python
# a)

def prep_lines_alph(p_text):
    p_text = p_text.lower()
    
    l_lines = p_text.split("\n")
    
    l_lines.sort()

    return l_lines

v_lines = prep_lines_alph(v_short_text)
print_list(v_lines)
```

```python
# b)

def get_max(sub):
 
    # returns maximum character
    if len(sub) >= 1:   
        return ord(sub[1])
    return 0
 

def prep_lines_ascii(p_text):
    p_text = p_text.lower()
    
    l_lines = p_text.split("\n")
    
    l_lines.sort(key=get_max)

    return l_lines

v_lines = prep_lines_ascii(v_short_text)
print_list(v_lines)

```

```python
# c)

def prep_lines(p_text):
    p_text = p_text.lower()
    
    l_lines = p_text.split("\n")
    
    l_lines.sort(reverse = True)

    return l_lines

v_lines = prep_lines(v_short_text)
print_list(v_lines)
```

```python
# d)

def prep_lines_ascii_r(p_text):
    p_text = p_text.lower()
    
    l_lines = p_text.split("\n")
    
    l_lines.sort(key=get_max, reverse = True)

    return l_lines

v_lines = prep_lines_ascii_r(v_short_text)
print_list(v_lines)
```

#### 4. Compute ngrams  

a. Find the 10 most common bigrams  
b. Find the 10 most common trigrams  

```python
# a)

def select_bigrams(p_text):
    p_text = p_text.replace("\n", " ")
    l_vowels = re.findall("\s[a-z]+\s[a-z]+\s", p_text)
    l_vowelcombos = []
    
    l_loop_count = 0
    for result in l_vowels:
        l_vowelcombos.append(l_vowels[l_loop_count])
        
        l_loop_count = l_loop_count + 1
        
    return l_vowelcombos  

v_output = create_unique_list(select_bigrams(v_short_text))
v_output.sort(reverse = True)

print_list(v_output[0:10])
```

```python
# b)

def select_trigrams(p_text):
    p_text = p_text.replace("\n", " ")
    l_vowels = re.findall("\s[a-z]+\s[a-z]+\s[a-z]+\s", p_text)
    l_vowelcombos = []
    
    l_loop_count = 0
    for result in l_vowels:
        l_vowelcombos.append(l_vowels[l_loop_count])
        
        l_loop_count = l_loop_count + 1
        
    return l_vowelcombos  

v_output = create_unique_list(select_trigrams(v_short_text))
v_output.sort(reverse = True)

print_list(v_output[0:10])
```
